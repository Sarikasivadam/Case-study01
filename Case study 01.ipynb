```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Read the dataset

```python
data1 = pd.read_csv("cars_data.csv")
data1
```

### 2. Check for the null values

```python
print(data1.isnull() .sum())
```

### 3. Plot a bar graph of male vs female buyers participated in the sales

```python
count=data1['Buyer Gender'].value_counts()
count
```

```python
fig1=plt.figure(figsize=(7,6))
count.plot.bar(title='male vs female buyers participated in the sales')
plt.xlabel("Buyer Gender")
plt.ylabel("Number of people")
```

### 4. Find the top 5 cars based on their sales price.


```python
sorted_sp=data1.sort_values(['Sale Price'],ascending=False)
```

```python
top5_cars=sorted_sp[["Make","Model","Sale Price"]].head()
top5_cars
```

### 5. Find the least 5 cars based on their Resell price.

```python
sorted_rp=data1.sort_values(["Resell Price"],ascending=True)
```

```python
least5_cars=sorted_rp[["Make","Model","Resell Price"]].head()
least5_cars
```
